```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units
```

```python
t = np.linspace(0,1,1000)*units.s
f = 1.*units.Hz
```

```python
# Plotting cosines of different phases
arg = (2.*np.pi*f*t).to(units.dimensionless_unscaled).value
plt.plot(t, np.cos(arg - 0), label='$\phi=0$')
plt.plot(t, np.cos(arg - np.pi/2), label='$\phi=\pi/2$')
plt.plot(t, np.cos(arg - np.pi), label='$\phi=\pi/2$')
plt.plot(t, np.cos(arg - 3*np.pi/2), label='$\phi=\pi/2$')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (arbitrary)')
plt.legend()
plt.show()
```

The imaginary number *i* in python is denoted 1j.

```python
print(np.power(1j, 2))
print(1j**2)
```

It can be used as the argument to a function.

```python
np.exp(1j)
```

```python
def cexp(phi):
    
    return np.exp(-2*np.pi*1j*phi)
```

```python
plt.plot(t, cexp(f*t).real, label='real')
plt.plot(t, cexp(f*t).imag, label='imag')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (arbitrary)')
plt.legend()
plt.show()
```

```python
plt.plot(cexp(f*t).real, cexp(f*t).imag)
plt.axis('equal')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```
